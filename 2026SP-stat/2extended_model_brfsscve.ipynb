This notebook covers the extension of base model with census data and (some) brfss data we thought might be relavant. Then we do some sensitivity analysis by removing Gaines County or other filtering outliers. We also tried simulating actual county level brfss data by fitting a linear model on phr level and then predicting county level medical cost percentage.

```r
# load lib
library(tidyverse)
county_data <- read.csv("data/county_data.csv")
census_data <- read.csv("data/texas_census.csv")
brfss_data <- read.csv("data/brfss_dummy.csv")
```

```r
county_data <- county_data %>%
  left_join(census_data, by = "County")

# # Verify merge worked
county_data %>%
  summarise(
    n_counties        = n(),
    missing_hispanic  = sum(is.na(pct_hispanic)),
    missing_poverty   = sum(is.na(pct_poverty)),
    missing_uninsured = sum(is.na(pct_uninsured))
  )
nacounty <- county_data[!complete.cases(county_data), ]
print(nacounty)
```

```r
library(lme4)
model_full <- glmer(
  outbreak ~
    # BRFSS PHR-level variables
    scale(medical_cost_pct) +
    scale(flu_shot_pct) +
    scale(cve) +
    # Census county-level proxies for suppressed BRFSS groups
    scale(pct_hispanic) +
    scale(pct_black) +
    scale(pct_poverty) +
    scale(pct_uninsured) +
    scale(pct_college) +
    
    # Offset
    offset(log(enrollment)) +
    
    # Random effects
    (1 | PHR),
    # (1 | PHR:County),
    
  family = poisson,
  data = county_data,
  nAGQ = 0
)
summary(model_full)
```

```r
install.packages("DHARMa")
library(DHARMa)

sim_res <- simulateResiduals(model_final, n = 1000)
plot(sim_res)                          # QQ + residual vs fitted
testOutliers(sim_res)                  # formal outlier test
testDispersion(sim_res)                # check if nbinom2 is enough

# Flag influential counties
outlier_idx <- outliers(sim_res)
county_data[outlier_idx, c("County", "outbreak", "population")]
```

```r
# Step 1: Find which rows were actually used in the model
model_rows <- as.integer(names(residuals(model_final, type = "pearson")))

# Step 2: Create a subset of county_data that matches the model
county_data_model <- county_data[model_rows, ]

# Step 3: Now safely attach residuals
county_data_model$resid <- residuals(model_final, type = "pearson")

# Step 4: Check which rows were dropped and why
dropped <- county_data[-model_rows, ]
print(dropped)                        # see what's special about rows 62, 160, 161, 162
colSums(is.na(county_data))           # find which columns have NAs
```

```r
# Piecewise weighting: 1 / 0.75 / 0.5
# bottom 90% -> 1
# 90-95%     -> 0.75
# top 5%     -> 0.5

p90 <- quantile(county_data$population, 0.90, na.rm = TRUE)
p95 <- quantile(county_data$population, 0.95, na.rm = TRUE)

county_data <- county_data %>%
  mutate(
    w_piece = case_when(
      population <= p90 ~ 1,
      population <= p95 ~ 0.75,
      TRUE ~ 0.5
    )
  )

model_weight <- glmmTMB(
  outbreak ~ scale(cve) + scale(pct_uninsured) +
    scale(pct_hispanic) + scale(pct_black) +
    scale(medical_cost_pct) +
    offset(log(population)) +
    (1 | PHR),
  family = nbinom2,
  data = county_data,
  weights = w_piece
)

summary(model_weight)

coef_piecewise <- summary(model_weight)$coefficients$cond

AIC(model_final, model_weight)
```

```r
# Winsorized weights:
# w = 1 / sqrt(min(population, cap))
# Try 90th, 95th, and 99th percentile caps

pop_cap_90 <- quantile(county_data$population, 0.90, na.rm = TRUE)
pop_cap_95 <- quantile(county_data$population, 0.95, na.rm = TRUE)
pop_cap_99 <- quantile(county_data$population, 0.99, na.rm = TRUE)

county_data_90 <- county_data %>%
  mutate(
    w_cap = 1 / sqrt(pmin(population, pop_cap_90))
  )

county_data_95 <- county_data %>%
  mutate(
    w_cap = 1 / sqrt(pmin(population, pop_cap_95))
  )

county_data_99 <- county_data %>%
  mutate(
    w_cap = 1 / sqrt(pmin(population, pop_cap_99))
  )

model_cap_90 <- glmmTMB(
  outbreak ~ scale(cve) + scale(pct_uninsured) +
    scale(pct_hispanic) + scale(pct_black) +
    scale(medical_cost_pct) +
    offset(log(population)) +
    (1 | PHR),
  family = nbinom2,
  data = county_data_90,
  weights = w_cap
)

model_cap_95 <- glmmTMB(
  outbreak ~ scale(cve) + scale(pct_uninsured) +
    scale(pct_hispanic) + scale(pct_black) +
    scale(medical_cost_pct) +
    offset(log(population)) +
    (1 | PHR),
  family = nbinom2,
  data = county_data_95,
  weights = w_cap
)

model_cap_99 <- glmmTMB(
  outbreak ~ scale(cve) + scale(pct_uninsured) +
    scale(pct_hispanic) + scale(pct_black) +
    scale(medical_cost_pct) +
    offset(log(population)) +
    (1 | PHR),
  family = nbinom2,
  data = county_data_99,
  weights = w_cap
)

summary(model_cap_90)
summary(model_cap_95)
summary(model_cap_99)

coef_unweighted   <- summary(model_final)$coefficients$cond
coef_weighted_90  <- summary(model_cap_90)$coefficients$cond
coef_weighted_95  <- summary(model_cap_95)$coefficients$cond
coef_weighted_99  <- summary(model_cap_99)$coefficients$cond

coef_unweighted
coef_weighted_90
coef_weighted_95
coef_weighted_99

AIC(model_final, model_cap_90, model_cap_95, model_cap_99)
```

```r
# Coefficient stability table

coef_unweighted_vec <- summary(model_final)$coefficients$cond[, "Estimate"]
coef_piecewise_vec  <- summary(model_weight)$coefficients$cond[, "Estimate"]
coef_cap_90_vec     <- summary(model_cap_90)$coefficients$cond[, "Estimate"]
coef_cap_95_vec     <- summary(model_cap_95)$coefficients$cond[, "Estimate"]
coef_cap_99_vec     <- summary(model_cap_99)$coefficients$cond[, "Estimate"]

coef_compare <- data.frame(
  Term = names(coef_unweighted_vec),
  Unweighted = coef_unweighted_vec,
  Piecewise = coef_piecewise_vec,
  Cap_90 = coef_cap_90_vec,
  Cap_95 = coef_cap_95_vec,
  Cap_99 = coef_cap_99_vec
)

coef_compare

pct_change <- coef_compare %>%
  mutate(
    Piecewise_pct_change = 100 * (Piecewise - Unweighted) / Unweighted,
    Cap_90_pct_change    = 100 * (Cap_90 - Unweighted) / Unweighted,
    Cap_95_pct_change    = 100 * (Cap_95 - Unweighted) / Unweighted,
    Cap_99_pct_change    = 100 * (Cap_99 - Unweighted) / Unweighted
  )

pct_change

coef_plot_df <- coef_compare %>%
  filter(Term != "(Intercept)") %>%
  pivot_longer(
    cols = -Term,
    names_to = "Model",
    values_to = "Estimate"
  )

ggplot(coef_plot_df, aes(x = Model, y = Estimate, color = Term, group = Term)) +
  geom_point(size = 3) +
  geom_line() +
  theme_minimal() +
  labs(
    title = "Coefficient Stability Across Weighted Models",
    y = "Coefficient Estimate",
    x = ""
  )
```
